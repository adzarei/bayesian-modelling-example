# Explore dataset and tools

```python
import pandas as pd

experiments = pd.read_csv('data/properMotions.csv')

display(experiments.head())

experiments.columns
```

```python
import numpy as np

measurements_1 = experiments[experiments['experiment'] == 1]

existing_sigma = measurements_1["sigma"].iloc[0]
calculated_sigma = np.std(measurements_1['y'])

print("Current sigma: ", existing_sigma)
print("Calculated sigma: ", calculated_sigma)

print("Equal: ", existing_sigma == calculated_sigma)
```

```python
import matplotlib.pyplot as plt


units = np.unique(experiments['experiment'])
plt.bar(range(len(units)), experiments['y'].groupby(experiments['experiment']).mean(), yerr=experiments['y'].groupby(experiments['experiment']).std())
plt.title("Mean of y by experiment and standard deviation")
plt.xlabel("Experiment")
plt.ylabel("Mean of y")
plt.show()
```

```python
plt.bar(range(len(units)), experiments['sigma'].groupby(experiments['experiment']).first())
plt.title("Sigma by experiment")
plt.xlabel("Experiment")
plt.ylabel("Sigma")
plt.show()
```
